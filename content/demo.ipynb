```python
import pandas as pd
import numpy as np

# Create a simple DataFrame
df = pd.DataFrame({
    'A': np.random.randn(5),
    'B': np.random.randn(5)
})
print(df)
```

```python
import matplotlib.pyplot as plt

# Simple plot
df.plot()
plt.title('Sample Plot')
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

# Load iris dataset and train a simple model
iris = load_iris()
clf = DecisionTreeClassifier()
clf.fit(iris.data[:100], iris.target[:100])
print(f"Model accuracy: {clf.score(iris.data[100:], iris.target[100:])}")
```
